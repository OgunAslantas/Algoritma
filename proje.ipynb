```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
veriler = pd.read_csv('odev_tenis.csv')
```

```python
# Kategorik veriyi numerik veriye dönüştürme:
play = veriler.iloc[:,-1:].values
windy = veriler.iloc[:,-2:-1].values

from sklearn import preprocessing

#le = preprocessing.LabelEncoder()
#play[:,-1] = le.fit_transform(veriler.iloc[:,-1])
#windy[:,-1] = le.fit_transform(veriler.iloc[:,-1])

veriler2 = veriler.apply(preprocessing.LabelEncoder().fit_transform)

# İlk kolona OneHotEncoder uygulanması gerekiyor:

c = veriler2.iloc[:,:1]
ohe = preprocessing.OneHotEncoder()
c = ohe.fit_transform(c).toarray()

# Verileri birleştirme:

havadurumu = pd.DataFrame(data=c,index=range(14),columns=['o','r','s'])
sonveriler = pd.concat([havadurumu,veriler.iloc[:,1:3]],axis=1)
sonveriler = pd.concat([sonveriler,veriler2.iloc[:,-2:]],axis=1)
```

```python
# Eğitim ve test için bölünmesi

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(sonveriler.iloc[:,:-1],sonveriler.iloc[:,-1:],test_size=0.40,random_state=0)

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train,y_train)

y_pred = regressor.predict(x_test)
y_pred
```
